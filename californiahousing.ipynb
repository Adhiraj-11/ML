```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from pandas.plotting import scatter_matrix
```

```python
ds = pd.read_csv('../input/californiamain/housing.csv')
ds.tail()
```

```python
housing_median_age = ds["housing_median_age"]
```

```python
housing_median_age.head()
```

```python
#1)How many blocks have a median housing age of 27?
counts = 0
for i in housing_median_age:
    if i == 27:
        counts += 1
print("number of houses with 27 as the median housing age are:",counts)
```

```python
#2)Add an attribute to the dataset that stores just the integer part of the latitude

int_latitude = ds['latitude'].fillna(0).astype(int)
```

```python
ds["int_latitude"] = int_latitude
```

```python
ds.head()
```

```python
set_s = set(ds['int_latitude'])

list_of_lats = list(set_s)

list_of_lats = sorted(list_of_lats)

cols = ['latitude_range', 'mean_population']
out_df = pd.DataFrame(columns=cols)

for latitude in list_of_lats:
    this_lat_df = ds.loc[ds['int_latitude'] == latitude]
    mean = this_lat_df['population'].mean()
    records = [(latitude, mean)]
    df_new = pd.DataFrame.from_records(records, columns = cols) 

    out_df = pd.concat([out_df, df_new])
out_df.head()
```

```python
#3)How many blocks have the total number of bedrooms between 500 and 1000?

bedrooms = ds["total_bedrooms"]
```

```python
counts_bedrooms = 0
for i in bedrooms:
    if i >= 500 and i <= 1000:
        counts_bedrooms += 1
print("No. of blocks having number of bedrooms between 500 and 1000 are: ", counts_bedrooms)
    
```
